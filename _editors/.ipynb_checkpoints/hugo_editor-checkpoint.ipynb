```python
import regex as re
```

```python
file_name = '/users/thomas/desktop/poetry/_raw/hugo_raw.txt'
```

```python
with open(file_name,'r+') as f:
    text = f.read()
```

```python
#text
```

```python
capture_title_expr = r'\n([A-Z][A-Z,\ ,\-,\',\:,\",\?]{4,})\.+\n'

title_expr = r'\n[A-Z][A-Z,\ ,\-,\',\:,\",\?]{4,}\.+\n'

front_matter, body = text.split('\n###\n')
#body = re.sub(capture_title_expr,r'\n\1\n',body)
```

```python
body[:100]
```

```python
body = re.sub(r'\[.*\]','',body)
```

```python
body = re.sub(r'\nPoems.*\n',r'',body)
body = re.sub(r'\n\ *\(?[A-Z]*\.\ [0-9]*\)?\.\n',r'',body)
```

```python
import numpy as np

titles = np.array([s.upper() for s in re.findall(capture_title_expr, body)])
poems = re.split(title_expr,body)[1:] # the first one is an empty string

len(titles), len(poems)
```

```python
for unique_title in np.unique(titles):
    
    title_index = np.where(titles==unique_title)[0]
    
    if not len(title_index) > 1:
        continue
        
    print(unique_title)
        
    if len(title_index) > 1:
        for ii,i in enumerate(title_index):
            titles[i] += ' ' + f'_{ii+1}_'
```

```python
titles
```

```python
final  = ''
final += front_matter + '\n###\n\n'

poems_dict = {}

for i, (title, poem) in enumerate(zip(titles,poems)):
        
    #poem = re.sub('\n(.)',r'\n\t\1',poem)
    adj_poem = re.sub(r'\n{4,}','\n\n\n',poem)
    if adj_poem[:2] == '\n\n':
        adj_poem = adj_poem[1:]
    if not adj_poem[0] == '\n':
        adj_poem = '\n' + adj_poem
    poems_dict[title.upper()] = adj_poem
    
    if len(adj_poem) > 10000:
        continue
        
    if len(adj_poem) < 100:
        assert False
    
    final += f'\n{title.upper()}\n'
    final += re.sub(r'\n{4,}','\n\n\n',adj_poem)[:-1]
    
    print(len(adj_poem))
```
