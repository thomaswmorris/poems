```python
import regex as re
```

```python
file_name = '/users/thomas/desktop/poetry/_raw/dickinson_raw.txt'
```

```python
with open(file_name,'r+') as f:
    text = f.read()
```

```python
capture_title_expr = r'\n[I,V,X,L]+\.\n\n([A-Z][A-ZÀ-Ú,\ ,\-,\:,\!,\']{3,})\.\n'

title_expr = r'\n[I,V,X,L]+\.\n\n[A-Z][A-ZÀ-Ú,\ ,\-,\:,\!,\']{3,}\.\n'

front_matter, body = text.split('\n###\n')
body = re.sub(r'\n([I,V,X,L]+\.)\n\n([A-Z][a-z, ])',r'\n\1\n\nUNTITLED.\n\n\2',body)
```

```python
#print(body[:10000])
```

```python
body = re.sub(r'[\[,\]]','',body)
```

```python
body = re.sub(r'\nPoems.*\n',r'',body)
body = re.sub(r'\n\ *\(?[A-Z]*\.\ [0-9]*\)?\.\n',r'',body)
```

```python
import numpy as np

titles = np.array([s.upper() for s in re.findall(capture_title_expr, body)])
poems = re.split(title_expr,body)[1:] # the first one is an empty string

len(titles), len(poems)
```

```python
for unique_title in np.unique(titles):
    
    title_index = np.where(titles==unique_title)[0]
    
    if not len(title_index) > 1:
        continue
        
    print(unique_title)
        
    if len(title_index) > 1:
        for ii,i in enumerate(title_index):
            titles[i] += ' ' + f'_{ii+1}_'
```

```python
titles
```

```python

final  = ''
final += front_matter + '\n###\n\n'

poems_dict = {}

for i, (title, poem) in enumerate(zip(titles,poems)):
        
    poem = poem.strip('\n')
    poems_dict[title.upper()] = poem
    
    if len(poem) < 100:
        print(title)


```

```python
text = ''

text += 'Emily Dickinson (1830-1886) [https://en.wikipedia.org/wiki/Emily_Dickinson]'
text += '\n\n###\n\n'

for title in titles:
    text += '~' + title + '~'
    text += '\n\n'
    text += poems_dict[title]
    text += '\n\n\n'
        
print(text[:400])
```

```python
import json

with open('../_text/dickinson.txt','w+') as f:
    f.write(text)
```
